```python
count = 0
for i in range(1, 10):
    if i % 2 == 0:
        count += 1
    elif i % 3 == 0:
        count -= 1
print(count)
```

```python
count = 0
for i in range(1, 6):
    if i % 2 == 0:
        for j in range(i):
            count += j
print(count)
```

```python
numbers = [1, 2, 3, 4, 5]
total = 0
for i in range(len(numbers)):
    if i % 2 == 0:
        continue
    total += numbers[i]
print(total)
```

```python
string = "Python"
result = ""
for i in range(len(string)):
    if i % 2 == 0:
        result += string[i].upper()
    else:
        result += string[i].lower()
print(result)
```

```python
numbers = [1, 2, 3, 4, 5]
doubled = [num * 2 for num in numbers if num % 2 != 0]
print(doubled)
```

```python
sentence = "Hello, world!"
vowels = [char for char in sentence if char.lower() in 'aeiou']

print(vowels)
```
